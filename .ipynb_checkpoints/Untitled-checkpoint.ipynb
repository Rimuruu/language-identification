```python
import csv
import re
import unicodedata
from string import punctuation
from string import digits
import copy
catalan = open("catalan--something--utf-8.txt", 'r', encoding='utf8')
#czech = open("czech--something--utf-8.txt", 'r', encoding='utf8')
danish = open("danish--something--utf-8.txt", 'r', encoding='utf8')
dutch = open("dutch--something--utf-8.txt", 'r', encoding='utf8')
english = open("english--narrative-of-a-gordon-pym--poe--utf-8.txt", 'r', encoding='utf8')
#esperanto = open("esperanto--something--utf-8.txt", 'r', encoding='utf8')
finnish = open("finnish--something--utf-8.txt", 'r', encoding='utf8')
french = open("french--la-recherche-de-labsolu--balzac--utf-8.txt", 'r', encoding='utf8')
german = open("german--der-prozess--kafka--utf-8.txt", 'r', encoding='utf8')
#hungarian = open("hungarian--something--utf-8.txt", 'r', encoding='utf8')
#icelandic = open("icelandic--something--utf-8.txt", 'r', encoding='utf8')
italian = open("italian--le-avventure-di-pinocchio--collodi--utf-8.txt", 'r', encoding='utf8')
latin = open("latin--something--cicero--utf-8.txt", 'r', encoding='utf8')
norwegian = open("norwegian--something--utf-8.txt", 'r', encoding='utf8')
#polish = open("polish--something--utf-8.txt", 'r', encoding='utf8')
portuguese = open("portuguese--something--utf-8.txt", 'r', encoding='utf8')
spanish = open("spanish--el-ingenioso-hidalgo-don-quijote-de-la-mancha--cervantes--utf-8.txt", 'r', encoding='utf8')
swedish = open("swedish--something--utf-8.txt", 'r', encoding='utf8')
#tagalog = open("tagalog--something--utf-8.txt", 'r', encoding='utf8')
#turkish = open("turkish--something--utf-8.txt", 'r', encoding='utf8')



cat = "catalan"
#cze = "czech"
dan = "danish"
dut = "dutch"
eng = "english"
#esp = "esperanto"
fin = "finnish"
fr = "french"
ger = "german"
#hun = "hungarian"
#ice = "icelandic"
ita = "italian"
lat = "latin"
nor = "norwegian"
#pol = "polish"
por = "portuguese"
spa = "spanish"
swe = "swedish"
#tag = "tagalog"
#tur = "turkish"

languages = [
    [catalan,cat,{}],
   # [czech,cze,{}],
    [danish,dan,{}],
    [dutch,dut,{}],
    [english,eng,{}],
  #  [esperanto,esp,{}],
    [finnish,fin,{}],
    [french,fr,{}],
    [german,ger,{}],
   # [hungarian,hun,{}],
   # [icelandic,ice,{}],
    [italian,ita,{}],
    [latin,lat,{}],
    [norwegian,nor,{}],
   # [polish,pol,{}],
    [portuguese,por,{}],
    [spanish,spa,{}],
    [swedish,swe,{}],
   # [tagalog,tag,{}],
   # [turkish,tur,{}],
]

languagesTrain = [

]

languagesTest = [

]

features={}

```

```python
#
# Ouverture du fichier source.
#
# D'après la documentation, le mode ''b'' est
# *obligatoire* sur les plate-formes où il est
# significatif. Dans la pratique, il est conseillé
# de toujours le mettre.
# L'ouverture est en écriture, d'où le ''w''.
#

def featureExist(x):
    for feature in features:
        if x in features:
            return True
    return False
    
def addFeature(x):
    if featureExist(x)==False:
        features[x] = 0

def iter3gram(x):
    for i in range(3, len(x)):
        addFeature(x[i-3:i])

def iter2gram(x):
    for i in range(2, len(x)):
        addFeature(x[i-2:i])
        
def replace_all(text, dic,char):
    for x in dic:
        text = text.replace(x, char)
    return text

def remove_all(text, dic):
    return replace_all(text, dic,' ')

specialChar = list(punctuation+'\n'+'«'+'»')
remove_digits = str.maketrans('', '', digits)
minLength = 50
nb = 170
nbTrain = int(nb*0.70);
nbTest = int(nb*0.30);
print(nbTrain)
fname = "data.csv"
with open(fname, 'w', newline='') as csvfile:
    for lang in languages:
        writer = csv.writer(csvfile,delimiter=' ')
        text = re.split("[!?.]+",lang[0].read()) # On découpe les textes en phrases.
        cleanText = list(filter(lambda y: y.strip(),list(map(lambda x : remove_all(x,specialChar),text)))) # .strip() filtre les textes vides
        minText = list(filter(lambda y: len(y)>minLength,cleanText)) # On veut que nos phrases ne soient pas trop petite.
        latinText =list(map(lambda x : unicodedata.normalize('NFKD', x).encode('ascii', 'ignore').decode("utf-8").lower(),minText))
        noNumberText = list((map(lambda x: x.translate(remove_digits), latinText)))
        for x in noNumberText[0:nbTrain]:
            iter3gram(x)
            iter2gram(x)
        newtext = [ x +";" + lang[1] +'\r\n' for x in noNumberText[0:nbTrain]] # unicodedata convertit les caractères spéciaux vers leur équivalent latin
        newtext2 = [ x +";" + lang[1] +'\r\n' for x in noNumberText[nbTrain:nbTrain+nbTest]] 
        langTrain = copy.copy(lang[1:])
        langTest = copy.copy(lang[1:])
        langTrain.append(len(newtext))
        langTrain.append(newtext)
        langTest.append(len(newtext2))
        langTest.append(newtext2)
        languagesTrain.append(langTrain)
        languagesTest.append(langTest)
        writer.writerow(newtext) 
    
    
    
def fitNgram(X):
    model = copy.copy(features)
    for key in features:
        if(len(key) == 3):
            if(X[1][key[0:2]]!=0):
                de = X[1][key[0:2]]
            else:
                de = 1
            model[key] = X[1][key]/de
    return model


    
for lang in languagesTrain:
        lang[1] = copy.copy(features)
        for key in features:
            for text in lang[3]:
                s = text.replace(";"+lang[0], "")
                lang[1][key] = s.count(key) + lang[1][key]
      
textesTest = []      
for lang in languagesTest:
    for text in lang[3]:
            t = []
            t.append(lang[0])
            t.append(copy.copy(features))
            t.append(text)
            for key in features:
                s = text.replace(";"+t[0], "")
                t[1][key] = s.count(key) + t[1][key]
            textesTest.append(t)
      
models = []

Xtest = []
Ytest = []
    
for lang in languagesTrain:
    models.append([lang,fitNgram(lang)])
    print("Nombre de textes pour ",lang[0]," : ",lang[2])

for text in textesTest:
    Xtest.append(fitNgram(text))
    Ytest.append(text[0])

print(len(text[1])) 
    
#print("Nos features : ", features)
print("Nombre de features : ",len(features))

#
# Fermeture du fichier source
#

csvfile.close()
```

```python
import numpy as np
import sklearn.metrics
import matplotlib.pyplot as plt
import matplotlib.cm as cm
def SVMLin(g,M):
    score = 0
    for f in M:
        score+= g[f]*M[f]
    return score

def predict(models,X):
    scores = []
    for lang in models:
        scores.append([lang[0][0],SVMLin(lang[1],X)])
    index = np.argmax(scores,axis=0)[1]
    return scores[index][0]
Ypred = []
for i in range(len(Xtest)):
    Ypred.append(predict(models,Xtest[i]))
    #print("Prediction pour ",Ytest[i]," on a ", predict(models,Xtest[i]))

print("Taux de précision : ",sklearn.metrics.accuracy_score(Ypred, Ytest))
confmat = sklearn.metrics.confusion_matrix(Ypred, Ytest)
plt.imshow(confmat)
plt.colorbar()
```
